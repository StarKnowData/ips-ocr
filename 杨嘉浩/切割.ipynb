#### 图片切割

```python
# 图片数
photo = 0

# 遍历json
for i in data:
    try:
        photo +=1
        # 找出图片名
        a = i['imageName']
        # 导入对应图片
        img = cv2.imread('./chushi/%s'%a)
        # 组件数
        num = 0
        # 找出xy所处位置并遍历
        for j in i['Data']['svgArr']:
            num +=1 
            # 记录需要用到的值
            x1 = int(j['data'][0:4:2][0]['x'])                 
            x2 = int(j['data'][0:4:2][1]['x'])
            y1 = int(j['data'][0:4:2][0]['y'])
            y2 = int(j['data'][0:4:2][1]['y'])
            # print(x1,x2,y1,y2)
            # 按位置切割
            img2 = img[y1:y2, x1:x2]
            # 保存
            cv2.imwrite('./output/%d_%d.jpg'%(photo,num), img2)
        print('good')
    except:
        pass      
```
